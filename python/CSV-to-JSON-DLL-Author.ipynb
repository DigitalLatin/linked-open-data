```python
import json, codecs, csv
```

```python
# Open the CSV containing the content.
infile = codecs.open('../csv/author-authorities-export.csv','r',encoding='utf-8')
author_data = csv.reader(infile,delimiter=',')

# Create an iterable for the script.
authors = [row for row in author_data]

# Remove the header row.
del authors[0]
```

```python
data = {}
data['@context']={}
data['@context']['@base'] = 'https://catalog.digitallatin.org/'
data['@context']['madsrdf'] = 'http://www.loc.gov/mads/rdf/v1#'
data['@context']['rdf'] = 'http://www.w3.org/1999/02/22-rdf-syntax-ns#'
data['@context']['dcterms'] = 'http://www.w3.org/1999/02/22-rdf-syntax-ns#'
data['@context']['Name'] = 'madsrdf:PersonalName'
data['@context']['AuthorizedName'] = 'madsrdf:authoritativeLabel'
data['@context']['Variant'] = 'madsrdf:hasVariant'
data['@context']['LatinVariant'] = 'madsrdf:variantLabel'
data['@context']['EnglishVariant'] = 'madsrdf:variantLabel'
data['@context']['NativeLanguageVariant'] = 'madsrdf:variantLabel'
data['@context']['AlsoKnownAs'] = 'madsrdf:variantLabel'
data['@context']['SpanishName'] = 'madsrdf:variantLabel'
data['@context']['FrenchName'] = 'madsrdf:variantLabel'
data['@context']['GermanName'] = 'madsrdf:variantLabel'
data['@context']['ItalianName'] = 'madsrdf:variantLabel'
data['@context']['ISNIName'] = 'madsrdf:variantLabel'
data['@context']['PerseusName'] = 'madsrdf:variantLabel'
data['@context']['Abbreviation'] = 'madsrdf:hasAbbreviationVariant'
data['@context']['Date'] = 'dcmi:date'
data['@context']['BirthDate'] = 'madsrdf:birthDate'
data['@context']['DeathDate'] = 'madsrdf:deathDate'
data['@context']['Floruit'] = 'madsrdf:temporal'
data['@context']['TimePeriod'] = 'dcterms:coverage'
data['@context']['ExactExternalAuthority'] = {}
data['@context']['ExactExternalAuthority']['@id'] = 'madsrdf:hasExactExternalAuthority'
data['@context']['ExactExternalAuthority']['@type'] = '@id'
data['@context']['VIAF'] = 'rdf:resource'
data['@context']['LCCN'] = 'rdf:resource'
data['@context']['LOC'] = 'rdf:resource'
data['@context']['BNE'] = 'rdf:resource'
data['@context']['BNF'] = 'rdf:resource'
data['@context']['DNB'] = 'rdf:resource'
data['@context']['ICCU'] = 'rdf:resource'
data['@context']['ISNI'] = 'rdf:resource'
data['@context']['Wikidata'] = 'rdf:resource'
data['@context']['WorldCat'] = 'rdf:resource'
data['@context']['Identifier'] = {}
data['@context']['Identifier']['@id'] = 'madsrdf:Identifier'
data['@context']['Identifier']['@type'] = '@id'
data['@context']['DLLid'] = 'madsrdf:idValue'
data['@context']['VIAFid'] = 'madsrdf:idValue'
data['@context']['LOCid'] = 'madsrdf:idValue'
data['@context']['STOAid'] = 'madsrdf:idValue'
data['@context']['PHIid'] = 'madsrdf:idValue'
data['@context']['CTS'] = 'madsrdf:idValue'
data['@graph'] = []

for i in authors:
    dll_link = i[0]
    authorized_name = i[1]
    native_language_name = i[2]
    # Function to process the author's Latin name, which may have multiple values.
    def latin_name_process():
        author_lat_name = i[3]

        # If there isn't a Latin name, return ''.
        if not author_lat_name:
            author_lat_name = ''
            return author_lat_name

        # If there is > 1 name, the names will be separated by ';'. Process the
        # names as a list, then return the joined list.
        elif ';' in author_lat_name:
            author_lat_names = []
            lat_name_list = author_lat_name.split(';')
            for name in lat_name_list:
                author_lat_names.append(name)
            author_lat_name = author_lat_names
            return author_lat_name

        # If there is only one name, return it.
        else:
            return author_lat_name

    latin_name = latin_name_process()

     # Function to process the author's English name, which might have multiple values.
    def english_name_process():
        author_english_name = i[4]
        # If there isn't an English name, return the metadata wrappers surrounding an empty value.
        if not author_english_name:
            author_english_name = ''
            return author_english_name

        # If there is > 1 name, the names will be separated by ';'. Process the
        # names as a list, then return the joined list.
        elif ';' in author_english_name:
            author_english_names = []
            english_name_list = author_english_name.split(';')
            for name in english_name_list:
                author_english_names.append(name)
            author_english_name = author_english_names
            return author_english_name

        # If there is only one name, return it.
        else:
            return author_english_name

    english_name = english_name_process()
    
    # Function to process the other alternative name forms, which might have multiple values.
    def author_alternative_name_process():
        author_alternative_name = i[5]
        # If aren't any alternative name forms, return an empty value.
        if not author_alternative_name:
            author_alternative_name = ''
            return author_alternative_name

        # If there is > 1 name, the names will be separated by ';'. Process the
        # names as a list, then return the joined list.
        elif ';' in author_alternative_name:
            author_alternative_names = []
            author_alternative_name_list = author_alternative_name.split(';')
            for name in author_alternative_name_list:
                author_alternative_names_list.append(name)
            author_alternative_name = author_alternative_names
            return author_alternative_name

        # If there is only one name, return it.
        else:
            return author_alternative_name

    alternative_name = author_alternative_name_process()
    
    # Process the name(s) in the field Related to DLL Authors, if there are any.
    def related_name_process():
        author_related_name = i[6]
        # If there aren't any values, return an empty value.
        if not author_related_name:
            author_related_name = ''
            return author_related_name

        # If there is > 1 name, the names will be separated by ';'. Process the
        # names as a list, then return the joined list.
        elif ';' in author_related_name:
            author_related_names = []
            related_name_list = author_related_name.split(';')
            for name in related_name_list:
                author_related_names.append(name)
            author_related_name = author_related_names
            return author_related_name

        # If there is only one name, return it.
        else:
            return author_related_name

    related_name = related_name_process()

    

    # Variant names

    bne_name = i[7]
    bnf_name = i[8]
    dnb_name = i[9]
    iccu_name = i[10]
    isni_name = i[11]
    perseus_name = i[12]
    abbreviation = i[13]
    variants = {'LatinVariant': latin_name,
               'EnglishVariant': english_name,
               'NativeLanguageVariant': native_language_name,
               'AlsoKnownAs': related_name,
               'AlternativeName': alternative_name,
               'SpanishName': bne_name,
               'FrenchName': bnf_name,
               'GermanName': dnb_name,
               'ItalianName': iccu_name,
               'ISNIName': isni_name,
               'PerseusName': perseus_name,
               'Abbreviation': abbreviation
               }

    # Dates
    author_bdate = i[14]
    author_ddate = i[15]
    author_floruit = i[16]
    time_period = i[17]
    dates = {'BirthDate': author_bdate,
    'DeathDate': author_ddate,
    'Floruit': author_floruit}

    # External authorities
    viaf_url = i[18]
    lccn_url = i[19]
    loc_url = i[20]
    bne_url = i[21]
    bnf_url = i[22]
    dnb_url = i[23]
    iccu_url = i[24]
    isni_url = i[25]
    wikidata_url = i[26]
    wikipedia_url = i[27]
    worldcat_url = i[28]
    authorities = {'VIAF': viaf_url, 
                   'LCCN': lccn_url,
                   'LOC': loc_url, 
                   'BNE': bne_url, 
                   'BNF': bnf_url,
                   'DNB': dnb_url,
                   'ICCU': iccu_url,
                   'ISNI': isni_url,
                   'Wikidata': wikidata_url,
                   'Wikipedia': wikipedia_url,
                   'Worldcat': worldcat_url}
    # Identifiers
    dll_id = i[29]
    viaf_id = i[30]
    loc_id = i[31]
    phi_id = i[32]
    stoa_id = i[33]
    cts_urn = i[34]
    identifiers = {'DLLid': dll_id,
                  'VIAFid': viaf_id,
                  'LOCid': loc_id,
                  'PHIid': phi_id,
                  'STOAid': stoa_id,
                  'CTS': cts_urn}
    authority = {'@id': dll_link,
    'Name':{
    'AuthorizedName': authorized_name,
    'Variant': variants,
    },
    'Date': dates,
    'TimePeriod': time_period,
    'ExactExternalAuthority': authorities,
    'Identifier': identifiers}

    data['@graph'].append(authority)
```

```python
with codecs.open('../json-ld/authors.json','a',encoding='utf-8') as outfile:
    json.dump(data,outfile,sort_keys = True, indent = 4, ensure_ascii = False)
```
